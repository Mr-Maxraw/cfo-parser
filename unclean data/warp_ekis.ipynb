```python
def get_type(row):
    if row['Краткое название'].rfind('ГБ') != -1:
        return 'Бюджетные учреждения'
    if row['Краткое название'].rfind('ГА') != -1:
        return 'Автономные учреждения'
    if row['Краткое название'].rfind('ГК') != -1:
        return 'Казенные учреждения'
    return 'Бюджетные учреждения'
```
