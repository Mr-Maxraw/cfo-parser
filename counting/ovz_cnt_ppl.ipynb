```python
def to_grp(row):
    ans = ['глухие', 'слабослышащие', 'слепые', 'слабовидящие', 'с нарушениями речи', 'с нарушениями опорно-двигательного аппарата', 'с задержкой психического развития', 'с расстройствами аутистического спектра', 'с умственной отсталостью', 'со сложной структурой дефекта', 'с туберкулезной интоксикацией', 'инвалиды (не по пмпк)']
    for i in range(1, 13):
        if row['inv'] in [str(i) + '(О)']:
            return ans[i - 1]
```

```python
res['new_inv'] = res.apply(lambda row: to_grp(row), axis=1)
```

```python
ans = res[['ekis', 'new_inv', 'sum']].groupby(['ekis', 'new_inv']).sum().unstack()
```
