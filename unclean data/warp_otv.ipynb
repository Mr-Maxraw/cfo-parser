```python
from openpyxl import Workbook, load_workbook
import pandas as pd
from openpyxl.utils.dataframe import dataframe_to_rows as df_to_row
```

```python
path_to_docs = 'C:/Users/PetukhovMD/Desktop/справочники/'
```

```python
def to_int(row):
    if pd.isna(row['Код ЕКИС']):
        return 0
    return int(row['Код ЕКИС'])
```

```python
base = pd.read_excel(path_to_docs + 'Ответственные.xlsx', 'Лист2')
```

```python
base
```

```python
ekis = pd.read_excel(path_to_docs + 'ИНН+ЕКИС.xlsx', 'Отчет')
```

```python
ekis
```

```python
res = pd.merge(ekis[['Код ЕКИС', 'ИНН']], base, how='right', left_on='ИНН', right_on='инн орг')
```

```python
res
```

```python
res = res.drop(columns=['ИНН'])
```

```python
res['Код ЕКИС'] = res.apply(lambda row: to_int(row), axis=1)
```

```python
res = res.fillna('-')
```

```python
wb = Workbook()
ws = wb.active
for row in df_to_row(res, header=True):
    ws.append(row)
wb.save('new_otv.xlsx')
```
