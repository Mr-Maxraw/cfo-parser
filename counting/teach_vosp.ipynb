```python
def for_szp(row, months):
    b = True
    for i in months:
        if pd.isna(row['stv_' + i]) :
            continue
        b = b & (row['day_' + i] == 1) & (row['stv_' + i] >= 1)
    return b
```

```python
all['for_szp'] = all.apply(lambda row: for_szp(row, months), axis=1)
```
