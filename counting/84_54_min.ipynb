```python
import pandas as pd
import numpy as np
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows as df_to_row
```

```python
from szp_funcs import create_res, print_df, path_to_data, load_groups, get_otv, get_fio
```

```python
ped, isp, aup = load_groups('input')
months = ['apr']
```

```python
otv = get_otv()
```

```python
fio = get_fio()
```

```python
res = create_res(months)
```

```python
min_teach = res[res['job_' + months[0]] == 'Учитель'][['inn', 'snils', 'sum_' + months[0], 'job_' + months[0], 'stv_' + months[0]]].groupby('snils').max().reset_index()
min_teach = min_teach[min_teach['sum_' + months[0]] < 84100]
```

```python
min_nped = res[res['job_' + months[0]].isin(ped) == False][['inn', 'snils', 'sum_' + months[0], 'job_' + months[0], 'stv_' + months[0]]].groupby('snils').max().reset_index()
min_nped = min_nped[min_nped['sum_' + months[0]] < 54500]
```

```python
donach = pd.read_excel(path_to_data + 'early_may' + '.xlsx')
```

```python
donach = donach.groupby(['inn', 'snils']).sum().reset_index()
```

```python
donach
```

```python
min_nped = pd.merge(min_nped, donach, on=['snils', 'inn'], how='left')
min_nped['sum_' + months[0]] = min_nped.sum_apr + min_nped['sum']
min_nped = min_nped[min_nped['sum_' + months[0]] < 54500][['inn', 'snils', 'sum_' + months[0], 'job_' + months[0], 'stv_' + months[0]]]
```

```python
min_nped
```

```python
min_nped_out = pd.merge(pd.merge(min_nped, fio, on='snils', how='left'), otv, how='left', on='inn')
```
