```python
import pandas as pd
import numpy as np
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows as df_to_row
```

```python
from szp_funcs import load_month, print_df, path_to_data, load_groups, get_otv, get_fio, gku_inn
```

```python
ped, isp, aup = load_groups('input')
months = ['may']
otv = get_otv()
fio = get_fio()
```

```python
res = load_month(months[0])
```

```python
res
```

```python
min_teach = res[(res['job_' + months[0]] == 'Учитель') & (res['day_' + months[0]] == 1) & (res['status_' + months[0]] == 'Работа') & (res['status_' + months[0]] == res['status_pref_' + months[0]]) & (res['stv_' + months[0]] >= 1) & (res['inn'].isin(gku_inn) == False)]
min_teach = min_teach[min_teach['sum_' + months[0]] < 84100]
```

```python
min_nped = res[(res['job_' + months[0]].isin(ped) == False) & (res['day_' + months[0]] == 1) & (res['status_' + months[0]] == 'Работа') & (res['status_' + months[0]] == res['status_pref_' + months[0]]) & (res['stv_' + months[0]] >= 1) & (res['inn'].isin(gku_inn) == False)]
min_nped = min_nped[min_nped['sum_' + months[0]] < 54500]
```

```python
min_nped_out = pd.merge(pd.merge(min_nped, fio, on='snils', how='left').drop_duplicates(['inn', 'snils']), otv, how='left', on='inn')
```

```python
min_teach_out = pd.merge(pd.merge(min_teach, fio, on='snils', how='left').drop_duplicates(['inn', 'snils']), otv, how='left', on='inn')
```
