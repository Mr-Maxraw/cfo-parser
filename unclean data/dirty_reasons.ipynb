```python
from openpyxl import Workbook, load_workbook
from openpyxl.utils.dataframe import dataframe_to_rows as df_to_row
import pandas as pd
```

```python
path_to_data = 'C:/Users/PetukhovMD/Desktop/февраль/'
path_to_docs = 'C:/Users/PetukhovMD/Desktop/справочники/'
```

```python
#Remove view restriction
pd.set_option('display.max_rows', 1500)
```

```python
df = pd.read_excel(path_to_data + 'ДОНМ_Начисления работников.xlsx', 'TDSheet')
```

```python
df
```

```python
def to_int(row):
    if pd.isna(row['КодЕКИС']):
        return 0
    return int(row['КодЕКИС'])
```

```python
df.КодЕКИС = df.apply(lambda row: to_int(row), axis=1)
```

```python
df_nan_work = df[(df.ДолжностьНаКонецМес.isna()) & (df.ДолжностьНачалоМес.isna())][['КодЕКИС', 'Организация', 'ОрганизацияИНН', 'СНИЛС', 'ФИО']]
df_nan_work
```

```python
df_bad_time = df[(df.НормаДней.isna()) | (df.НормаДней > 30)][['КодЕКИС', 'Организация', 'ОрганизацияИНН', 'СНИЛС', 'ФИО']]
df_bad_time
```

```python
df_podr = df[df.ВидЗанятостиНачалоМес == 'Подработка'][['КодЕКИС', 'Организация', 'ОрганизацияИНН', 'СНИЛС', 'ФИО']]
df_podr
```

```python
df_to_row(df_podr, header=True).__next__()
```

```python
df_status = df[(df.СостояниеНачалоМес.isna()) & (df.ВидЗанятостиНачалоМес == 'Основное место работы')][['КодЕКИС', 'Организация', 'ОрганизацияИНН', 'СНИЛС', 'ФИО']]
df_status
```

```python
df1 = pd.read_excel(path_to_data + 'ДОНМ_Кадровые_паспортные данные.xlsx', 'TDSheet')
```

```python
df1.КодЕКИС = df1.apply(lambda row: to_int(row), axis=1)
df1
```

```python
df_nosnils = df1[(df1.СНИЛС.isna())][['КодЕКИС', 'Организация', 'ИННорг', 'СНИЛС', 'ФИО']]
df_nosnils
```

```python
df_nan_work_pd = df1[df1.Должность.isna()][['КодЕКИС', 'Организация', 'ИННорг', 'СНИЛС', 'ФИО']]
df_nan_work_pd
```

```python
wb = Workbook()
ws = wb.active
ws.append(['Тип ошибки', 'КодЕКИС', 'Организация', 'ОрганизацияИНН', 'СНИЛС', 'ФИО'])
for row in df_to_row(df_nan_work_pd, header=False):
    if len(row) < 6:
        continue
    if row[1] in ['КодЕКИС', None]:
        continue
    ws.append(['Нет должности']+row[1:])
for row in df_to_row(df_nosnils, header=False):
    if len(row) < 6:
        continue
    if row[1] in ['КодЕКИС', None]:
        continue
    ws.append(['Нет снилса']+row[1:])
wb.save('test2.xlsx')
```

```python
wb = Workbook()
ws = wb.active
ws.append(['Тип ошибки', 'КодЕКИС', 'Организация', 'ОрганизацияИНН', 'СНИЛС', 'ФИО'])
for row in df_to_row(df_podr, header=True):
    #'Вид занятости - "Подработка"'
    if len(row) < 6:
        continue
    if row[1] in ['КодЕКИС', None]:
        continue
    ws.append(['Вид занятости - "Подработка"']+row[1:])
for row in df_to_row(df_bad_time, header=False):
    #'Вид занятости - "Подработка"'
    if len(row) < 6:
        continue
    if row[1] in ['КодЕКИС', None]:
        continue
    ws.append(['Ошибка в норме дней']+row[1:])
for row in df_to_row(df_nan_work, header=False):
    #'Вид занятости - "Подработка"'
    if len(row) < 6:
        continue
    if row[1] in ['КодЕКИС', None]:
        continue
    ws.append(['Отсутствует должность']+row[1:])
for row in df_to_row(df_status, header=False):
    #'Вид занятости - "Подработка"'
    if len(row) < 6:
        continue
    if row[1] in ['КодЕКИС', None]:
        continue
    ws.append(['Отстствует статус']+row[1:])
wb.save('test.xlsx')
```
