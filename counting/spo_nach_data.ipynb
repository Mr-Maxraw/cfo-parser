```python
inns = sch_types[sch_types.sch_type == 'СПО']['inn'].to_list()
wb = Workbook()
for month in months:
    nach = pd.read_excel(path_to_data + month + '_nach.xlsx')
    nach = nach[nach.inn.isin(inns)]
    kadr = pd.read_excel(path_to_data + month + '_kadr.xlsx')
    kadr = kadr[kadr.inn.isin(inns)]
    ans = pd.merge(kadr[['inn', 'snils', 'status', 'job', 'type']].drop_duplicates(), nach, how='left', on=['inn', 'snils', 'job', 'type'])
    ws = wb.create_sheet()
    ws.title = month
    for i in df_to_row(ans, header=True, index=False):
        ws.append(i)
wb.save(r'C:\Users\PetukhovMD\Desktop\СПО начисления' + '\\' + 'начисления по СПО с ' + months[0] + ' по ' + months[-1] + '.xlsx')
```
