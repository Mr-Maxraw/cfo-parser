```python
df = pd.DataFrame(cnt, columns=['name', 'inn', 'subtype', 'pers_inn'])
df
```

```python
def get_type(row):
    if row['name'].rfind('ГБ') != -1:
        return 'Бюджетные учреждения'
    if row['name'].rfind('ГА') != -1:
        return 'Автономные учреждения'
    if row['name'].rfind('ГК') != -1:
        return 'Казенные учреждения'
    return 'Бюджетные учреждения'
```

```python
df['type'] = df.apply(lambda row: get_type(row), axis=1)
df
```
