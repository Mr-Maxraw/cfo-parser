```python
from szp_funcs import print_df, path_to_docs, path_to_data, to_double
```

```python
all = pd.read_excel(path_to_docs + 'pay_type.xlsx', sheet_name='Все').rename(columns={'Вид ': 'type', 'Наименование ': 'name'})[['type', 'name']]
```

```python
look_up = all['name'].to_list()
```

```python
inst_pe = pd.read_excel(path_to_data + 'inst_pe.xlsx').rename(columns={'СНИЛС': 'snils'})
```

```python
inst_pe_snils = inst_pe['snils'].to_list()
```

```python
teach_pe = pd.read_excel(path_to_data + 'teach_pe.xlsx').rename(columns={'СНИЛС': 'snils'})
```

```python
teach_pe_snils = teach_pe['snils'].to_list()
```

```python
dec_nach = pd.read_excel(path_to_data + 'dec_nach_21.xlsx')
```

```python
dec_nach
```

```python
names_dec = pd.DataFrame(dec_nach['nach_name'].unique()).rename(columns={0: 'nach_name'})
```

```python
names_dec
```

```python
def matcher(row):
    ans = process.extractOne(row['nach_name'], look_up, scorer=fuzz.WRatio, score_cutoff=60)
    if ans:
        return ans[0]
    return ans
```

```python
names_dec['new_name'] = names_dec.apply(lambda row: matcher(row), axis=1)
```

```python
names_dec[names_dec.new_name.isna()]
```

```python
names_dec = pd.merge(names_dec, all, how='left', left_on='new_name', right_on='name')[['nach_name', 'type']]
```

```python
names_dec
```

```python
dec_nach = pd.merge(dec_nach, names_dec, how='left')
```

```python
dec_nach[(dec_nach.snils.isin(inst_pe_snils)) & (dec_nach.type == 'Стимулирующие выплаты')][['snils', 'nach_sum']].groupby('snils').sum().reset_index()['nach_sum'].mean()
```

```python
dec_nach[(dec_nach.snils.isin(teach_pe_snils)) & (dec_nach.type == 'Стимулирующие выплаты')][['snils', 'nach_sum']].groupby('snils').sum().reset_index()['nach_sum'].mean()
```
